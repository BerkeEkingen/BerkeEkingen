```python
import numpy as np
import pandas as pd

powers = np.array([
    [base ** power for power in range(10)]
    for base in range(1, 11)
])

print(powers)
```

```python
df = pd.DataFrame(
    powers,
    index=[f"base_{i}" for i in range(1, 11)],
    columns=[f"pow_{i}" for i in range(10)]
)

print(df)
```
