```python
import pandas as pd
import numpy as np

s = pd.Series(
    data=range(1, 101),
    index=[f"e{i:03d}" for i in range(1, 101)]
)

print(s)
```

```python
powers = np.array([
    [i**j for j in range(10)]
    for i in range(1, 11)
])

print(powers)

df = pd.DataFrame(
    powers,
    index=[f"base_{i}" for i in range(1,11)],
    columns=[f"pow_{j}" for j in range(10)]
)

print(df)
```

```python
print(np.array_equal(df.values, powers))
```

```python
print(df.dtypes)

df_small = df.astype(np.uint16)

print(df_small.dtypes)
```

```python
iris = pd.read_csv("iris.csv")

iris.head()
```

```python
iris.columns = [
    col.replace(" ", "_")
    for col in iris.columns
]

iris.head(10)
```

```python
print(iris.memory_usage(deep=True))

iris_small = iris.copy()

for col in iris_small.select_dtypes(include=np.number).columns:
    iris_small[col] = pd.to_numeric(
        iris_small[col],
        downcast="float"
    )

print(iris_small.memory_usage(deep=True))
```

```python
iris_small.to_clipboard(index=False)

print("Copied to clipboard.")
```
