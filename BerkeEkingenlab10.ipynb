```python
pop2023 = demo[
    (demo["TIME_PERIOD"] == 2023) &
    (demo["indic_de"] == "AVG")
]

top20 = pop2023.nlargest(20, "OBS_VALUE")

px.pie(
    top20,
    names="geo",
    values="OBS_VALUE",
    title="Top 20 Population Structure 2023"
).show()
```

```python
pl_pop = pop2023[
    pop2023["geo"] == "PL"
]["OBS_VALUE"].iloc[0]

pop2023["difference"] = abs(
    pop2023["OBS_VALUE"] - pl_pop
)

closest = (
    pop2023[pop2023["geo"] != "PL"]
    .sort_values("difference")
    .head(3)
)

print(closest[["geo", "OBS_VALUE"]])
```
