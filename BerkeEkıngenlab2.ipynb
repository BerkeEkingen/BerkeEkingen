```python
import numpy as np

arr = np.arange(2, 34, 2).reshape(4, 4)

print("Array:")
print(arr)

print("\nData type:", arr.dtype)
print("Number of elements:", arr.size)
print("Size of one element (bytes):", arr.itemsize)
print("Allocated memory (bytes):", arr.nbytes)
```

```python
view_arr = arr.view(np.uint8)

print("View array:")
print(view_arr)

print("\nIs it a view?")
print(view_arr.base is arr)

print("\nData type:", view_arr.dtype)
print("Number of elements:", view_arr.size)
print("Size of one element (bytes):", view_arr.itemsize)
print("Allocated memory (bytes):", view_arr.nbytes)
```

```python
view_center = arr[1:3, 1:3]

copy_center = arr[1:3, 1:3].copy()

print("View:")
print(view_center)

print("\nCopy:")
print(copy_center)

print("\nView base:")
print(view_center.base is arr)

print("Copy base:")
print(copy_center.base is arr)
```

```python
arr4 = np.tile(np.arange(1, 5), (4, 1))

mirror = arr4[:, ::-1]

result = np.concatenate((arr4, mirror), axis=1)

print(result)
```

```python
Z_5 = np.ones((2, 2), dtype=int)

Z1 = Z_5 * 1
Z2 = Z_5 * 2
Z3 = Z_5 * 3
Z4 = Z_5 * 4

top = np.concatenate((Z1, Z2), axis=1)
bottom = np.concatenate((Z3, Z4), axis=1)

result = np.concatenate((top, bottom), axis=0)

print(result)
```
