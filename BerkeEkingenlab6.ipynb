```python
import pandas as pd
import numpy as np

orders = pd.read_csv("orders.csv")

# 1.1
print(orders.isna().sum())

# 1.2
idx = np.random.choice(orders.index, int(len(orders) * 0.10), replace=False)
orders.loc[idx, "Seller"] = "MISSING"

# 1.3
idx = np.random.choice(orders.index, int(len(orders) * 0.05), replace=False)
orders.loc[idx, "OrderID"] = np.nan

# 1.4
idx = np.random.choice(orders.index, int(len(orders) * 0.20), replace=False)
orders.loc[idx, "OrderDate"] = np.nan

# 1.5
idx = np.random.choice(orders.index, int(len(orders) * 0.15), replace=False)
orders.loc[idx, "OrderValue"] = np.nan

# 1.6
orders_missing = orders.copy()
```

```python
# 2.1
orders_missing["OrderID"] = orders_missing["OrderID"].fillna(0)

# 2.2
orders_missing["OrderDate"] = orders_missing["OrderDate"].ffill()

# 2.3
orders_missing["OrderValue"] = (
    orders_missing.groupby("Country")["OrderValue"]
    .transform(lambda x: x.fillna(x.mean()))
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
orders["OrderValue"].hist()
plt.title("Original Dataset")
plt.show()

plt.figure(figsize=(8,5))
orders_missing["OrderValue"].hist()
plt.title("After Filling Missing Values")
plt.show()
```

```python
orders.groupby("Country")["OrderValue"].mean()
```

```python
orders["OrderDate"] = pd.to_datetime(orders["OrderDate"])

orders.groupby(
    [orders["OrderDate"].dt.year,
     orders["OrderDate"].dt.month]
)["OrderValue"].mean()
```

```python
orders.groupby(
    [orders["OrderDate"].dt.year, "Seller"]
).size()
```

```python
revenue = orders.groupby(
    ["Country", "Seller"]
)["OrderValue"].sum()

revenue.groupby(level=0).idxmax()
```

```python
quarter_revenue = orders.groupby(
    [orders["OrderDate"].dt.year,
     orders["OrderDate"].dt.quarter]
)["OrderValue"].sum()

quarter_revenue.sort_values(ascending=False)
```
