```python
import numpy as np

a_small_one = np.arange(1, 65).reshape(8, 8)

print(a_small_one)
```

```python
print("Sum of each column:")
print(np.sum(a_small_one, axis=0))

print("\nCumulative sum for each row:")
print(np.cumsum(a_small_one, axis=1))

print("\nAverage value for each row:")
print(np.mean(a_small_one, axis=1))
```

```python
even_elements = a_small_one[a_small_one % 2 == 0]

print(even_elements)
```

```python
no_even_small_one = a_small_one.astype(float).copy()

no_even_small_one[no_even_small_one % 2 == 0] = np.nan

print(no_even_small_one)
```

```python
print("Sum for each row:")
print(np.nansum(no_even_small_one, axis=1))

print("\nAverage for each column:")
print(np.nanmean(no_even_small_one, axis=0))

print("\nSmallest value:")
print(np.nanmin(no_even_small_one))
```

```python
restored_array = np.where(
    np.isnan(no_even_small_one),
    a_small_one,
    no_even_small_one
)

print(restored_array)
```

```python
normal_100 = np.random.normal(
    loc=0,
    scale=1,
    size=(100, 100)
)

print(normal_100)
```

```python
row_means = np.mean(normal_100, axis=1)
column_means = np.mean(normal_100, axis=0)

print("Row means:")
print(row_means)

print("\nColumn means:")
print(column_means)

row_diff = np.mean(np.abs(row_means))
column_diff = np.mean(np.abs(column_means))

print("\nAverage absolute difference of row means from 0:")
print(row_diff)

print("\nAverage absolute difference of column means from 0:")
print(column_diff)
```

```python
import timeit

big_array = np.random.normal(size=(1000, 1000))
```

```python
txt_time = timeit.timeit(
    lambda: np.savetxt("array.txt", big_array),
    number=1
)

npy_time = timeit.timeit(
    lambda: np.save("array.npy", big_array),
    number=1
)

npz_time = timeit.timeit(
    lambda: np.savez_compressed("array.npz", big_array),
    number=1
)

print("TXT save:", txt_time)
print("NPY save:", npy_time)
print("NPZ save:", npz_time)
```
