```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def radar_chart(df, players, attributes):

    N = len(attributes)

    angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))

    for player in players:
        row = df[df["Name"] == player].iloc[0]

        values = row[attributes].tolist()
        values += values[:1]

        ax.plot(angles, values, linewidth=2, label=player)
        ax.fill(angles, values, alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(attributes)

    ax.set_ylim(0, 100)

    plt.legend()
    plt.show()
```

```python
fifa = pd.read_csv("FIFA22_official_data.csv")

attrs = [
    "Crossing",
    "HeadingAccuracy",
    "ShortPassing",
    "Volleys",
    "Dribbling",
    "Curve",
    "FKAccuracy",
    "LongPassing",
    "BallControl",
    "Acceleration"
]

players = ["M. Sabitzer", "J. Cuadrado"]

radar_chart(fifa, players, attrs)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

df = pd.read_csv("pkn.txt")

fig, ax = plt.subplots()

def update(frame):

    ax.clear()

    data = df.iloc[:20 + frame]

    ax.plot(data["TIME"], data["CLOSE"])

    ax.set_xlabel("Time")
    ax.set_ylabel("Close Price")

ani = FuncAnimation(
    fig,
    update,
    frames=len(df)-20,
    interval=100
)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("pkn.txt")

plt.figure(figsize=(12,6))

plt.plot(df["TIME"], df["CLOSE"])

min_idx = df["CLOSE"].idxmin()
max_idx = df["CLOSE"].idxmax()

plt.annotate(
    "MIN",
    (df.loc[min_idx,"TIME"], df.loc[min_idx,"CLOSE"])
)

plt.annotate(
    "MAX",
    (df.loc[max_idx,"TIME"], df.loc[max_idx,"CLOSE"])
)

plt.show()
```

```python
import pandas as pd

df = pd.read_csv("pkn.txt")

df["DATE"] = pd.to_datetime(df["DATE"])

result = df.groupby(
    [df["DATE"].dt.year,
     df["DATE"].dt.month]
)["CLOSE"].agg(["min","max"])

def highlight_min(s):
    return [
        "background-color: yellow"
        if v == s.min()
        else ""
        for v in s
    ]

def highlight_max(s):
    return [
        "background-color: lightgreen"
        if v == s.max()
        else ""
        for v in s
    ]

result.style \
    .apply(highlight_min, subset=["min"]) \
    .apply(highlight_max, subset=["max"])
```
