```python
import numpy as np

arr1 = np.arange(1, 11)
print(arr1)

```

```python
arr2 = arr1.astype(np.float32)
print(arr2)
```

```python
arr3 = 2 ** np.arange(2, 17)
print(arr3)
```

```python
arr4 = np.arange(1, 26).reshape(5, 5)

print(arr4)
print("\nLast value:")
print(arr4[-1, -1])

print("\nSecond column:")
print(arr4[:, 1])

print("\nThird row:")
print(arr4[2, :])

print("\nMain diagonal:")
print(np.diag(arr4))

print("\nValues 9,10,14,15:")
print(arr4[1:3, 3:5])
```

```python
arr5 = np.arange(1, 28).reshape(3, 3, 3)

print("Original array:")
print(arr5)

print("\ny = 1:")
print(arr5[:, 1, :])

print("\nz = 1 and z = 3:")
print(arr5[:, :, [0, 2]])
```

```python
import time
import numpy as np

my_list = list(range(10000))

start = time.time()
result = []
for x in my_list:
    result.append(x * 2)
print("For loop:", time.time() - start)

start = time.time()
result = [x * 2 for x in my_list]
print("List comprehension:", time.time() - start)

np_array = np.arange(10000)

start = time.time()
result = np_array * 2
print("NumPy:", time.time() - start)
```
